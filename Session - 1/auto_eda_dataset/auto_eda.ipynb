# Import Modules

```python
import pandas as pd
import dtale
```

# Load Dataset

```python
# Set the dataset file path.
dataset_path = "./store_sales_data_v2.csv"

# Read data from the directory using pandas.
dataset = pd.read_csv(dataset_path)
```

# Initiatie Dtale

```python
# Pass dataset to the show method in dtale.
dtale_show_instance = dtale.show(dataset)

# Use the open method browser of the dtale view object to open the dataset in a browser.
dtale_show_instance.open_browser()
```

```python
# You can also use the dtale show method without assigning it to see the data inside the notebook.
# dtale.show(dataset)
```

# Manage Running Instances.

```python
# Display running instances of dtale.
dtale.instances()
```

```python
# Fetch an instance.
dtale_instance = dtale.get_instance(1)

# Kill the instance.
dtale_instance.kill()
```
